```python
from dataclasses import dataclass
from typing import Optional, Tuple
from sqlalchemy.engine import URL
from sqlalchemy.types import String
from sqlalchemy.orm import Session
from sqlalchemy import select, insert, MetaData, Table

metadata = MetaData()

@dataclass
class DltTable:
    name: str
    incremental_column: Optional[str] = None
    primary_key: Optional[Tuple[str]] = None
    dedup: bool = True
    initial_value: Optional[str] = None
    alias: Optional[str] = None

    def __post_init__(self):
        if self.alias is None:
            self.alias = self.name
TABLES = [
    DltTable(
        name="model_has_roles",
        primary_key=("role_id", "model_id", "model_type")
    ),
    DltTable(
        name="roles",
        primary_key=("id",),
        incremental_column="updated_at",
        alias="type_of_roles"
    )
]

doris_engine = create_engine(
    URL.create(
        drivername="doris+pymysql",
        username=os.getenv(f'DORIS_{ENV}_USER'),
        password=os.getenv(f'DORIS_{ENV}_PASSWORD'),
        host=os.getenv(f'DORIS_{ENV}_HOST'),
        port=int(os.getenv('DORIS_PORT')),
        database=os.getenv(f'DORIS_{ENV}_NAME'),
    ),
    pool_pre_ping=True,
    pool_recycle=3600
)

for table in TABLES:
    source = fs_source(
        bucket_url="s3://livrina-dev-raw/to_load_doris_dataset",
        credentials=minio_cred,
        file_glob=table.name + '/' + "*.jsonl",
        incremental=dlt.sources.incremental(
            "modification_date",
            range_end="open"
        )
    ) | read_jsonl()

    data = []
    for data_ in source:
        data.append(data_)
    
    df = pd.DataFrame({'data' : data})
    df['data'] = df['data'].apply(json.dumps)
    """
    df.to_sql(
        name=table.alias,
        con=doris_engine,
        if_exists='append',
        index=False,
        dtype={'data': String}
    )
    """

    table_obj = Table(
        table.alias,
        metadata,
        autoload_with=doris_engine
    )

    with Session(doris_engine) as s:
        q = insert(table_obj)
        s.execute(q, df.to_dict(orient='records'))
        s.commit()
    
```

```python
import boto3
from datetime import datetime, timezone

def list_objects_modified_after(
    bucket: str,
    prefix: str,
    modified_after: datetime,
    endpoint_url: str = None,
    access_key: str = None,
    secret_key: str = None
):
    """
    Liste les objets modifiés après `modified_after` dans un bucket S3/MinIO.
    """
    s3 = boto3.client(
        "s3",
        endpoint_url=endpoint_url,
        aws_access_key_id=access_key,
        aws_secret_access_key=secret_key,
        config=boto3.session.Config(signature_version="s3v4"),
        region_name="us-east-1"
    )

    paginator = s3.get_paginator("list_objects_v2")

    results = []
    for page in paginator.paginate(Bucket=bucket, Prefix=prefix):
        for obj in page.get("Contents", []):
            if obj["LastModified"] > modified_after:
                results.append({
                    "key": obj["Key"],
                    "size": obj["Size"],
                    "last_modified": obj["LastModified"]
                })

    return results


if __name__ == "__main__":
    # Exemple d'usage

    MINIO_ENDPOINT = os.getenv('MINIO_ENDPOINT')  # Ton endpoint MinIO
    ACCESS_KEY = os.getenv('MINIO_AIRFLOW_ACCESS_KEY')                    # tes credentials
    SECRET_KEY = os.getenv('MINIO_AIRFLOW_SECRET_KEY') 

    BUCKET = "livrina-dev-raw"
    PREFIX = "to_load_doris_dataset/model_has_roles/"  # dossier cible

    # Filtrer tous les fichiers modifiés après cette date
    modified_after = datetime(2025, 1, 14, 0, 0, 0, tzinfo=timezone.utc)

    objs = list_objects_modified_after(
        bucket=BUCKET,
        prefix=PREFIX,
        modified_after=modified_after,
        endpoint_url=MINIO_ENDPOINT,
        access_key=ACCESS_KEY,
        secret_key=SECRET_KEY
    )

    print("Fichiers modifiés récemment :")
    for o in objs:
        print(o)
```
